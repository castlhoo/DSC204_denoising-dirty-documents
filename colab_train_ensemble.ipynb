# Denoising Dirty Documents — Ensemble Training on Colab GPU

**Setup:**
1. Upload `train_sean.py` and `ensemble_predict.py` to Colab
2. Upload `data.zip` (compressed data folder) to Colab
3. Make sure Runtime > Change runtime type > **T4 GPU**
4. Run all cells in order

```python
# Step 1: Check GPU
import torch
print('CUDA available:', torch.cuda.is_available())
print('GPU:', torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'None')
```

```python
# Step 2: Extract data
import zipfile, os

# Adjust path if needed
zip_path = '/content/data.zip'
extract_path = '/content/'

with zipfile.ZipFile(zip_path, 'r') as z:
    z.extractall(extract_path)

print('Contents:', os.listdir('/content/data'))
```

```python
# Step 3: Train model with seed=42
!python /content/train_sean.py \
    --data-dir /content/data \
    --epochs 80 \
    --batch-size 32 \
    --patches-per-image 8 \
    --patch-size 64 \
    --base-channels 48 \
    --patience 15 \
    --seed 42 \
    --model-path /content/model_seed42.pth \
    --output-csv /content/pred_seed42.csv
```

```python
# Step 4: Train model with seed=123
!python /content/train_sean.py \
    --data-dir /content/data \
    --epochs 80 \
    --batch-size 32 \
    --patches-per-image 8 \
    --patch-size 64 \
    --base-channels 48 \
    --patience 15 \
    --seed 123 \
    --model-path /content/model_seed123.pth \
    --output-csv /content/pred_seed123.csv
```

```python
# Step 5: Train model with seed=456
!python /content/train_sean.py \
    --data-dir /content/data \
    --epochs 80 \
    --batch-size 32 \
    --patches-per-image 8 \
    --patch-size 64 \
    --base-channels 48 \
    --patience 15 \
    --seed 456 \
    --model-path /content/model_seed456.pth \
    --output-csv /content/pred_seed456.csv
```
